```python
from astropy.io import fits
from astropy.wcs import WCS
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Assuming you have a FITS file named 'example.fits'
fits_file = './pid_1_118/VISP_2022_06_03T20_32_03_522_00630205_I_BZWKJ_L1.fits'

hdul = fits.open(fits_file)
hdul.info()
```

```python
header = hdul[1].header
header
```

```python
hdul[1].data.shape
```

```python
image = hdul[1].data[0,:,:]
image.shape
image[0,:].shape
```

```python
# Plot each row
plt.figure(figsize=(10, 6))

plt.imshow(image)

plt.title('title')
plt.xlabel('slit')
plt.ylabel('wavelength')
plt.grid(False)
plt.show()
```

```python

plt.figure(figsize=(10, 6))


plt.plot(image[:,1000])

plt.title('title')
plt.xlabel('wavelength')
plt.ylabel('intensity')
plt.grid(False)
plt.show()
```
